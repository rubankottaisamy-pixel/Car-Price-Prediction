```python
sns.countplot(x="Fuel_Type",data=cardf)
plt.AbstractContextManager
```

```python
sns.barplot(data=cardf, x='Fuel_Type', y='Selling_Price', estimator='mean',palette='Blues')
plt.title('avg selling price by fuel type')
plt.ylabel('price(laksh)')
plt.show()
```

```python
sns.countplot(data=cardf, x='Fuel_Type' , palette='Set2')
plt.title('Number of Listing by Fuel type')
plt.show()
```

```python
sns.boxplot(data=cardf, x='Fuel_Type', y='Selling_Price', palette='pastel')
plt.title('Price Distribution by Fuel Type')
plt.show()
```

```python
sns.histplot(data=cardf, x='Selling_Price' ,bins=30 , kde=True, palette='pastel')
plt.title('Distribution od Selling Price')
plt.show()
```

```python
numeric_cardf = cardf[['Selling_Price','Present_Price',
                  'Kms_Driven','Year']]
corr = numeric_cardf.corr()

sns.heatmap(corr, annot=True, fmt='.2f',
            cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(cardf[['Selling_Price','Present_Price',
                  'Kms_Driven','Year']],
             diag_kind='kde')
plt.show()
```
